# **We will build a classification model which can classify these three species based on the length and width measurements of their sepals and petals.**

```python
# importing libraries
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Iris.csv")
df.head()
```

```python
# scatterplot
colors = ['red', 'orange', 'blue']
species = ['virginica','versicolor','setosa']
```

```python
for i in range(3):
    x = df[df['species'] == species[i]]
    plt.scatter(x['sepal_length'], x['sepal_width'], c = colors[i], label=species[i])
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.legend()
```

```python
for i in range(3):
    x = df[df['species'] == species[i]]
    plt.scatter(x['petal_length'], x['petal_width'], c = colors[i], label=species[i])
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.legend()
```

```python
for i in range(3):
    x = df[df['species'] == species[i]]
    plt.scatter(x['sepal_length'], x['petal_length'], c = colors[i], label=species[i])
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.legend()
```

```python
for i in range(3):
    x = df[df['species'] == species[i]]
    plt.scatter(x['sepal_width'], x['petal_width'], c = colors[i], label=species[i])
plt.xlabel("Sepal Width")
plt.ylabel("Petal Width")
plt.legend()
```

We can clearly see from the pairplot that setosa can be linearly classified and where as versicolor and virginia data points overlap each other.
From the picture also we can clearly see the difference between setosa and other two species.

The average sepalLength value of each species can be inferred from here.For setosa-5cm ,versicolor-5.8,virginical-6.3.

```python
from sklearn.model_selection import train_test_split
# train - 90
# test - 10
X = df.drop(columns=['species'])
Y = df['species']
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.10)
```

```python
x_train
```

```python
x_test
```

# **Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

model1 = LogisticRegression()
# model training
model1.fit(x_train, y_train)

y_pred = model1.predict(x_test)
cm = confusion_matrix(y_test,y_pred)

# print metric to get performance
print('Confusion matrix: \n',cm)
print("Accuracy: ",model1.score(x_test, y_test))
print('Classification report: \n',classification_report(y_test,y_pred))
```

```python
# visualize with seaborn library
sns.heatmap(cm,annot=True,fmt="d")
```

```python
from sklearn.model_selection import cross_val_score

print("cross validation:",cross_val_score(model1,x_train,y_train,cv=8,scoring='accuracy').mean())
```

# **knn - k-nearest neighbours**

```python
from sklearn.neighbors import KNeighborsClassifier
model2 = KNeighborsClassifier(n_neighbors=4)
model2.fit(x_train, y_train)

y_pred = model2.predict(x_test)
cm = confusion_matrix(y_test,y_pred)

# print metric to get performance
print('Confusion matrix: \n',cm)
print("Accuracy: ",model2.score(x_test, y_test))
print('Classification report: \n',classification_report(y_test,y_pred))
```

```python
# visualize with seaborn library
sns.heatmap(cm,annot=True,fmt="d")
```

```python
print("cross validation:",cross_val_score(model2,x_train,y_train,cv=8,scoring='accuracy').mean())
```

```python
# visualize with seaborn library
sns.heatmap(cm,annot=True,fmt="d")
```
